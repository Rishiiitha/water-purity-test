```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('water_potability_smote.csv')
```

```python
for col in df.columns[:-1]:
    sns.kdeplot(data=df, x=col, hue="Potability")
    plt.title(f"Distribution of {col} by Potability",color="blue")
    plt.show()
```
